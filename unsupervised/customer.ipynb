```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample customer data: [Age, Spending Score]
data = {
    "Age": [25, 30, 22, 40, 35, 23, 45, 34, 55, 60],
    "Spending Score": [70, 65, 78, 20, 25, 80, 15, 28, 10, 12]
}
df = pd.DataFrame(data)

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=2, random_state=0)
df["Cluster"] = kmeans.fit_predict(df[["Age", "Spending Score"]])

# Add cluster centers to DataFrame
centers = kmeans.cluster_centers_
centers_df = pd.DataFrame(centers, columns=["Age", "Spending Score"])

# 🎨 Plot with Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Age", y="Spending Score", hue="Cluster", palette="Set2", s=100)
sns.scatterplot(data=centers_df, x="Age", y="Spending Score", marker='X', s=200, color='red', label="Centroid")

plt.title("Customer Segmentation with K-Means Clustering")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
